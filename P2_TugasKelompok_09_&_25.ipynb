```python
# 1. Akses file image lokal (KTM) dan tampilkan menggunakan OpenCV dan Matplotlib
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow


img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/PCVK25_3F_09/KTM_Evan.jpg')
ktm_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(ktm_img)
```

```python
tinggi = ktm_img.shape[0]
lebar = ktm_img.shape[1]
img = ktm_img.copy()

# Menutup Barcode
start_barcode = (int(lebar*0.80), int(tinggi*0.50))
end_barcode   = (int(lebar*0.98), int(tinggi*0.80))
cv2.rectangle(img, start_barcode, end_barcode, (255, 0, 0), -1)

# Menutup Kota "Malang"
start_text = (int(lebar*0.21), int(tinggi*0.08))
end_text   = (int(lebar*0.30), int(tinggi*0.13))
cv2.rectangle(img, start_text, end_text, (0, 0, 255), -1)

final_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

cv2_imshow(final_img)
```
