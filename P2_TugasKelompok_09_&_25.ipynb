```python
import cv2
import matplotlib.pyplot as plt

# Membaca gambar
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/PCVK25_3F_09/KTM_Evan.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Salin gambar
edit_img = img_rgb.copy()

# Tutup area logo, teks BNI, dan QR dengan kotak hitam
cv2.rectangle(edit_img, (325, 0), (325+266, 0+116), (0,0,0), -1)   # Logo
cv2.rectangle(edit_img, (541, 39), (541+30, 39+77), (0,0,0), -1)   # Teks BNI
cv2.rectangle(edit_img, (480, 233), (480+81, 233+76), (0,0,0), -1) # QR Code

# Tampilkan hasil
plt.imshow(edit_img)
plt.axis('off')
plt.title('KTM Setelah Ditutup (Kelompok 10)')
plt.show()
```
